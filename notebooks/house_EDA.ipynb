```python
binary_cols = ['mainroad','guestroom','basement','hotwaterheating','airconditioning','prefarea']
```

df_copy = pd.get_dummies(

    columns=['furnishingstatus'],
    dtype=int
)

```python
df_copy.drop(columns=['furnishingstatus_semi-furnished'],inplace=True)
```
